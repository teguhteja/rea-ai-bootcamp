## Challenge 1
### Menentukan Apakah Sebuah Bilangan Merupakan Bilangan Prima

Kamu diminta membuat fungsi yang memeriksa apakah sebuah bilangan adalah bilangan prima. Bilangan prima adalah bilangan yang hanya bisa dibagi oleh 1 dan dirinya sendiri. Fungsi ini akan menggunakan **loop** dan **conditional** untuk memeriksa setiap kemungkinan pembagi dari bilangan tersebut.

```python
def is_prime(num):
    # Memeriksa apakah num adalah bilangan prima
    for i in range(2,num):
      if num % i == 0:
        return False

    return True

print(is_prime(7)) # -> True (karena 7 hanya bisa dibagi oleh 1 dan 7)
print(is_prime(8)) # -> False (karena 8 bisa dibagi oleh 1, 2, 4, dan 8)
```

## Challenge 2
### Menghitung Bilangan Fibonacci ke-N
Soal ini meminta kamu untuk membuat fungsi yang mengembalikan bilangan Fibonacci ke-`n`.



```python
!pip install ipdb
```

```python
import ipdb

def fibonacci(n):
    # Menghitung bilangan Fibonacci ke-n
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    a, b = 0, 1
    for _ in range(2, n + 1):
      a, b = b, a + b
    return b

print(fibonacci(10)) # -> 55
print(fibonacci(15)) # -> 610
```

## Challenge 3
### Memutasi Daftar Berdasarkan Kondisi (FizzBuzz)
Kamu harus membuat fungsi yang mengubah daftar angka sesuai dengan aturan **FizzBuzz**. Jika angka adalah kelipatan 3, ganti dengan "**Fizz**". Jika kelipatan 5, ganti dengan "**Buzz**". Jika kelipatan 15, ganti dengan "**FizzBuzz**". Daftar diubah langsung menggunakan **loop** dan **if-elif-else**.

```python
def mutate_list(lst):
    # Memutasi daftar sesuai aturan FizzBuzz
    any = []
    for i in lst :
      s = ''
      if i % 3 == 0:
        s += "Fizz"
      if i % 5 == 0:
        s +="Buzz"
      if i % 3 != 0 and i % 5 != 0 :
        s = i
      any.append(s)

    return any

print(mutate_list([3, 5, 15, 7, 9])) # -> ['Fizz', 'Buzz', 'FizzBuzz', 7, 'Fizz']
```

## Challenge 4
### Memeriksa Bilangan Armstrong
Soal ini meminta kamu untuk membuat fungsi yang memeriksa apakah sebuah bilangan adalah bilangan Armstrong. Bilangan Armstrong adalah bilangan yang jumlah digitnya dipangkatkan dengan jumlah digit itu sendiri, menghasilkan bilangan aslinya. Misalnya, 153 adalah bilangan Armstrong karena $1^{3} + 5^{3} + 3^{3} = 153$

```python
def is_armstrong(n):
    # Memeriksa apakah n adalah bilangan Armstrong
    n_str = str(n)
    any = 0
    for i in n_str:
      any += int(i) ** len(n_str)
    return n == any

print(is_armstrong(153)) # -> True (1^3 + 5^3 + 3^3 = 153)
print(is_armstrong(123)) # -> False
print(is_armstrong(1634)) # -> True (1^4 + 6^4 + 3^4 + 4^4 = 1634)
```

## Challenge 5
### Mengelompokkan Kata Berdasarkan Panjang
Kamu diminta membuat fungsi yang menerima daftar kata dan mengelompokkan kata-kata tersebut berdasarkan panjangnya ke dalam dictionary. Setiap panjang kata menjadi key dalam dictionary, dan daftar kata-kata yang memiliki panjang tersebut menjadi value.

```python
def group_by_length(words):
    # Mengelompokkan kata berdasarkan panjangnya
  word_lengths = {}  # Inisialisasi dictionary kosong
  for word in words:
    length = len(word)  # Dapatkan panjang kata
    if length not in word_lengths:
      word_lengths[length] = []  # Buat list kosong jika panjang kata belum ada
    word_lengths[length].append(word)  # Tambahkan kata ke list dengan panjang yang sesuai
  return word_lengths


print(group_by_length(["cat", "dog", "mouse", "elephant", "ant"]) )
# -> {3: ["cat", "dog", "ant"], 5: ["mouse"], 8: ["elephant"]}
```

## Challenge 6
### Mencari Sublist Terpanjang yang Terdiri dari Elemen-Elemen Berurutan
Kamu diminta untuk mencari sublist terpanjang dari daftar yang terdiri dari elemen-elemen berurutan (**bertambah atau berkurang dengan selisih 1**).

```python
def longest_consecutive_sublist(lst):
  # Mencari sublist terpanjang yang terdiri dari elemen-elemen berurutan
  if not lst:  # Handle kasus list kosong
      return []

  longest_sublist = []
  current_sublist = [lst[0]]  # Inisialisasi dengan elemen pertama

  for i in range(1, len(lst)):
      if abs(lst[i] - lst[i - 1]) == 1:  # Periksa selisih 1
          current_sublist.append(lst[i])
      else:
          if len(current_sublist) > len(longest_sublist):
              longest_sublist = current_sublist
          current_sublist = [lst[i]]  # Mulai sublist baru

  # Periksa sublist terakhir
  if len(current_sublist) > len(longest_sublist):
      longest_sublist = current_sublist

  return longest_sublist

print(longest_consecutive_sublist([2, 3, 4, 5, 6, 8, 9, 11, 12, 13])) # -> [2, 3, 4, 5, 6]
print(longest_consecutive_sublist([4, 3, 2, 1, 6, 7, 8])) # -> [4, 3, 2, 1]
```

## Challenge 7
### Menghitung Karakter dengan Frekuensi Tertinggi

Kamu harus membuat fungsi yang menerima sebuah string dan mengembalikan karakter dengan frekuensi tertinggi. Jika terdapat beberapa karakter dengan frekuensi yang sama, kembalikan karakter yang pertama kali muncul dalam string.

```python
from collections import Counter

def longest_char_freq(sentence):
    # Hitung frekuensi karakter
    counter = Counter(sentence)
    # Temukan karakter dengan frekuensi tertinggi
    max_count = max(counter.values())
    for char in sentence:  # Iterasi sesuai urutan asli
        if counter[char] == max_count:
            return char

print(longest_char_freq("hello world")) # -> 'l'
print(longest_char_freq("testcase")) # -> 't'
```

### Aplikasi Sederhana untuk Menghitung Jumlah Karakter dan Kata
Soal ini mengajak kamu untuk membuat aplikasi dengan Gradio. Aplikasi menerima sebuah teks, lalu mengembalikan jumlah karakter dan jumlah kata dalam teks tersebut. Gradio digunakan untuk membuat antarmuka grafis aplikasi ini.

```python
import gradio as gr

def text_stats(text):
    stats = {
        "Characters": len(text),  # Panjang string (termasuk spasi)
        "Words": len(text.split())  # Jumlah kata berdasarkan spasi
    }
    return stats

# Contoh penggunaan:
# Input: "Hello world"
# Output: {"Characters": 11, "Words": 2}
gr.Interface(fn=text_stats, inputs="text", outputs="text").launch()
```
